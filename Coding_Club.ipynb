```python
# Program for Randomly Generating Passwords
import random
import string
UppercaseLetters = string.ascii_uppercase
LowercaseLetters = string.ascii_lowercase
Digits = string.digits
All = UppercaseLetters + LowercaseLetters + Digits


def Random_Password():
    Sampling = random.choices(All,k=8)
    Password = ''.join(Sampling)
    print("The Random Password is ", Password)
    
Random_Password()
    
```

```python
# Program to Plot World Population throughout the Ages
import matplotlib.pyplot as plt
import csv
def World_Population():
  
  x=[]
  y=[]

  with open('Population.csv', 'r') as csvfile:
    plots= csv.reader(csvfile, delimiter=',')
    for row in plots:
        x.append(int(row[0]))
        y.append(int(row[1]))

  plt.plot(x,y,marker='o')
  plt.title('Data from CSV File: World Population Growth')
  plt.xlabel('Year')
  plt.ylabel('Population')
  plt.show()
  

World_Population()

```
